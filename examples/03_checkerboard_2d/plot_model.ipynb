```python
import pygmt
import h5py
import numpy as np
```

```python
with h5py.File('OUTPUT_FILES/final_model_ph.h5', 'r') as f:
    print(f.keys())
    svel = f['vel'][:]
    x = f['lon'][:]
    y = f['lat'][:]
    periods = f['period'][:]
xx, yy = np.meshgrid(x, y, indexing='ij')
```

```python
idx_period = 5
region = [x.min(), x.max(), y.min(), y.max()]
fig = pygmt.Figure()
fig.basemap(region=region, projection='M4i', frame=True)
grid = pygmt.surface(z=svel[idx_period,:,:].ravel(), x=xx.ravel(),y=yy.ravel(), region=region, spacing='0.01/0.01')
fig.grdimage(grid=grid, cmap='seis')
fig.show()
```
